```python
# Исходные данные
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
# Функции
# Ищем введенный пользователем № документа в имеющихся документах
def _find_owner(num_doc):
  for el in documents:
    if num_doc == el['number']:
      return el
  else:
      return None

# Функция выполняет действие с полученным результатом: если документ есть в имеющихся документах => выводим имя владельца,
#  если нет, то выводим 'Документ не найден в базе'
def find_owner(num_doc):
  doc = _find_owner(num_doc)
  if doc is None:
    print('Документ не найден в базе') 
  else:
    print('Владелец документа:', doc['name'])         

# Ищем номер полки по номеру документа    
def _find_shelf(num_doc):
  for shelf, number in directories.items():
    if num_doc in number:
      return shelf
  else:
    return None    

# Функция выполняет действие с полученным результатом: если документ есть => выводим № полки, если нет, то
# то выводим 'Документ не найден в базе'
def find_shelf(num_doc):
  number_shelf = _find_shelf(num_doc)
  if number_shelf is None:
    print('Документ не найден в базе') 
  else:
    print('Документ хранится на полке:', number_shelf) 

# Функция выводит всю имеющуюся информацию о документах
def full_info():
  for  el in documents:
    print(f"№: {el['number']}, тип: {el['type']}, владелец: {el['name']}, полка хранения: {_find_shelf(el['number'])}")

# Функция проверяет, существует ли в данный момент полка (по ключу)
def exists_shelf(shelf):
  return str(shelf) in directories.keys()

# Функция добавляет номер полки в имеющийся перечень полок (словарь)
def add_shelf(shelf):
  global directories
  directories[str(shelf)] = []

# Функция выводит актуальную информацию о полках (их номера)
def list_shelf():
  return 'Текущий перечень полок: '+ ", ".join(directories.keys())

# Функция выводит сообщение и актуальную информацию о полках (их номера)
def view_message(message):
  print(message, list_shelf())

# Функция проверяет наполнение полки (пустая или нет)
def is_free_shelf(shelf):
  if exists_shelf(shelf):
    return len(directories[str(shelf)]) == 0 
  return True

# Функция удаляет полку
def delete_shelf(shelf):
  del directories[str(shelf)]
```

```python
# Функции для выполнения команд
def comand_p():
  user_input = input('Введите номер документа: ')
  find_owner(user_input)

def comand_s(): 
  user_input = input('Введите номер документа: ')
  find_shelf(user_input) 

def comand_l(): 
  full_info() 

def comand_ads():  
  shelf_input = input('Введите номер новой полки: ')
  is_exists = exists_shelf(shelf_input)
  if not is_exists:
    add_shelf(shelf_input)
    view_message("Полка добавлена.")
  else:
    view_message("Такая полка уже существует!")  

def comand_ds():  
  shelf_input = input('Введите номер полки: ')
  is_exists = exists_shelf(shelf_input)
  is_free = is_free_shelf(shelf_input)
  if is_exists and is_free:
    delete_shelf(shelf_input)
    view_message("Полка удалена.")
  else:
    view_message("На полке есть документы, удалите их перед удалением полки.") if is_exists else view_message("Такой полки не существует.") 
```

```python
# Программа
while True:
  command = input('Введите команду: ')
  if command == 'p': 
    comand_p()
  elif command == 's': 
    comand_s()
  elif command == 'l':
    comand_l()
  elif command == 'ads':
    comand_ads()
  elif command == 'ds':
    comand_ds()
  elif command == 'q': 
    break
  else:
    print('Команда отсутствует. Введите новую')
```
