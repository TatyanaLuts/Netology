```python
# Задание 2
queries = [
'смотреть сериалы онлайн бесплатно',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про американский футбол',
]

result = {}
count = 0
for el in queries:
    quantity = len(el.split()) # считаем количество слов в каждом запросе
    if quantity in result.keys():
        result[quantity] = result[quantity]+1
    else: 
        result[quantity] = 1
count = sum(result.values())

for k, v in list(result.items()):
    print(f'Поисковых запросов, содержащих {k} слов(а): {v*100/count:.2f}%')
```
