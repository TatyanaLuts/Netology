```python
# Задание 3
boy_list = ['Ivan', 'Petr', 'Aleksey', 'Pavel']
girl_list = ['Olga', 'Anna', 'Vera', 'Svetlana']
if len(boy_list) == len(girl_list):
    s_boy_list = sorted(boy_list)
    s_girl_list = sorted(girl_list)
    pair = zip(s_boy_list, s_girl_list)
    print('Идеальные пары:')
    for el in zip(s_boy_list, s_girl_list):
        print(el[0], 'и', el[1])
else:
    print('Внимание, кто-то может остаться без пары!')
```

```python
# Задание 4
countries_temperature = [
['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['Germany', [57.2, 55.4, 59, 59, 53.6]],
['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

summa = 0.0
average = 0.0
cel = 0.0
print('Средняя температура в странах:')
for country in countries_temperature:
    summa = sum(country[1])
    average = summa/len(country[1])
    cel = round((average-32)*5/9,1)
    print(country[0], '-', cel, 'C')
```
